```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("Mall_Customers.csv")
data
```

```python
from sklearn.cluster import KMeans
```

```python
x=data.drop(['CustomerID','Gender','Age'],axis=1)
#print(x)
y=x.values
print(y)
```

```python
#from sklearn.preprocessing import MinMaxScaler
#r=MinMaxScaler().fit(y)
#r.transform(y)
```

```python
wcss=[]
for i in range(1,11):
    KM=KMeans(n_clusters=i,init='k-means++')
    KM.fit(y)
    wcss.append(KM.inertia_)
```

```python
#plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss)
plt.show
```

```python
KM1=KMeans(n_clusters=5,init='k-means++')
#KM1.fit(data)'
```

```python
label=KM1.fit_predict(y)
print(label)
```

```python
a=data['Annual Income (k$)']
b=data['Spending Score (1-100)']
plt.scatter(a,b,c=label)
centroid=KM1.cluster_centers_
cent_a=centroid[:,0]
cent_b=centroid[:,1]
plt.scatter(cent_a,cent_b,s=100,marker='D')
plt.show
```
